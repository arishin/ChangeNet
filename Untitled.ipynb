```python
from pathlib import Path
import re
from os import listdir
import glob



def splitData(trainfile, trainout):

    nlist = []

    with trainfile.open() as f:
        s = f.readline()
        nlist.extend(s.split())

    f.close()

    flist = [ "{0:0>3}".format(int(x)) for x in nlist] # list of 3 digit numbers used to get training files

    rawpath = Path('VL-CMU-CD-dataset')/ 'raw'

    with open(trainout, "w") as ofile:
        for ndir in flist:

            RGBpath = rawpath / ndir / 'RGB'
            GTpath = rawpath / ndir / 'GT'

            onlyfiles = [ f for f in glob.glob(str(RGBpath)+'//1*.png')]

            fileNumber = [ re.search('1_([0-9]+)\.png', xf).group(1) for xf in onlyfiles]
            fileNumber.sort()

            # [Target, Ref, GT]

            trainItem = [ str(RGBpath / ("2_"+str(fileNumber[0])+".png")) , str(RGBpath / ("1_"+str(fileNumber[0])+".png")) , str(GTpath / ("gt"+str(fileNumber[0])+".png")) ]

            pLine = trainItem[0]+" "+ trainItem[1]+ " " + trainItem[2] + "\n"
            ofile.write(pLine)
            print(pLine)

    ofile.close()  
    
trainfile = Path('VL-CMU-CD') / 'train_split.txt'
trainout = Path('train.txt')

valfile =   Path('VL-CMU-CD') / 'test_split.txt'
valout = Path('val.txt')

splitData (trainfile, trainout)
splitData (valfile, valout)
```
